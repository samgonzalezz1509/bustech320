```python
num1 = float(input("Enter the first number: "))
operator = input("Enter an operator (+, -, *, /): ")
num2 = float(input("Enter the second number: "))

if operator == '+':
    result = num1 + num2
elif operator == '-':
    result = num1 - num2
elif operator == '*':
    result = num1 * num2
elif operator == '/':
    if num2 != 0:
        result = num1 / num2
    else: print("Error: Division by zero is not allowed.")

print(f"Result: {result}")
```

```python
def getValues(n):
    assert isinstance(n, int)
    alist = []
    print(f"Please enter {n} weights:")
    for i in range(n):
        while True:
            try:
                val = float(input(f"Measure {i + 1}: "))
                alist.append(val)
                break
            except ValueError:
                print("Invalid input. Please enter a numeric weight.")
    return alist

def calcAverage(alist):
    assert isinstance(alist,list)
    try:
      return sum(alist[2:]) / len(alist[2:])
    except:
      raise Exception("Error: You must enter more than 2 measures to calculate an adjusted average.")



def main():
    num_measures = int(input("How many total measures are you recording? "))

    if num_measures <= 2:
        print("Error: You must enter more than 2 measures to calculate an adjusted average.")
    else:
        weights = getValues(num_measures)
        result = calcAverage(weights)

        print("\n--- Statistics ---")
        print(f"Total measures recorded: {len(weights)}")
        print(f"Measures used (excluding first two): {len(weights) - 2}")
        print(f"Calculated Average Weight: {result:.2f}")

if __name__ == "__main__":
    main()
```

```python
class Person:
    def __init__(self, first_name, last_name):
        self.first_name = first_name
        self.last_name = last_name

    def display(self, formal=True):
        if formal:
            print(f"{self.first_name} {self.last_name}")
        else:
            print(f"{self.first_name}")

    def __str__(self):
        return f"Person Object: {self.first_name} {self.last_name}"

name = Person("John", "Doe")
print(name)
name.display()
name.display(formal=False)
```

```python
class Student:
  def __init__(self, name, number):
    self.__name = name
    self.number = number
    self.__courses = []
  def get_name(self):
    return self.__name

  def set_name(self, newName):
    self.__name = newName

  def __len__(self):
    return len(self.__courses)

  def get_courses(self):
    return self.__courses

  def enroll(self, newCourse):
    if newCourse not in self.__courses:
      self.__courses.append(newCourse)
    else:
      print(f"Student {self.__name} is already enrolled in {newCourse}")
```

```python
import math

class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def distance(self, other):
    dx = self.x - other.x
    dy = self.y - other.y
    return math.sqrt(dx ** 2 + dy ** 2)

class Rectangle:
  def __init__(self):
    self.p1 = Point(0,0)
    self.p2 = Point(3,0)
    self.p3 = Point(0,3)
    self.p4 = Point(0,4)

    self.width = self.p1.distance(self.p2) # equivalent to : Point.distance(self.p1, self.p2)
    self.height = self.p1.distance(self.p3)

  def area(self):
    return self.width * self.height






```

```python
class Employee:
  def __init__(self, name, hourly_wage, hours_worked):
    self.name = name
    self.__hourly_wage = hourly_wage
    self.hours_worked = hours_worked
    print("Employee created")

  def calculate_pay(self):
    return self.__hourly_wage * self.hours_worked

  def get_hourly_wage(self):
    return self.__hourly_wage

  def set_hourly_wage(self, new_hourly_wage):
    if self.__hourly_wage <= 0:
      print("Negative wage values not allowed.")
    else:
      self.__hourly_wage = new_hourly_wage

  def display_info(self):
    print(f"Employee: {self.name}\nHourly wages: {self.hourly_wage}")

class Manager(Employee):
  def __init__(self, name, hourly_wage, hours_worked, bonus):
    super().__init__(name, hourly_wage, hours_worked)
    self.bonus = bonus

  def calculate_pay(self):
    return (1+self.bonus) * self.get_hourly_wage() * self.hours_worked

employee =Employee("Bob", 50, 10)
manager = Manager("Alice", 100, 15, 10)


print(employee)
print(manager)
print(f"Employee pay: {employee.calculate_pay()}")
print(f"Manager pay: {manager.calculate_pay()}")
```
