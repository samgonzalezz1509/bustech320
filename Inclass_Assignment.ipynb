```python
sales = float(input("Set sales value($)"))
cogs = 0.4 * sales
gross_profit= sales - cogs
tax_rate = float(input("Set tax rate value(%)"))
taxes = gross_profit * tax_rate / 100
net = gross_profit - taxes

print(f"Sales: {sales}$")
print(f"Cost of Goods Sold: {cogs}$")
print(f"Gross Profit: {gross_profit}$")
print(f"Taxes: {taxes}$")
print(f"Net Income: {net}$")
```

```python
num1 = float(input("Enter the first number: "))
operator = input("Enter an operator (+, -, *, /): ")
num2 = float(input("Enter the second number: "))

if operator == '+':
    result = num1 + num2
elif operator == '-':
    result = num1 - num2
elif operator == '*':
    result = num1 * num2
elif operator == '/':
    if num2 != 0:
        result = num1 / num2
    else: print("Error: Division by zero is not allowed.")

print(f"Result: {result}")
```

```python
print("Hour Fee")
hour = 1
while hour < 9:
  fee = 5 + 2.5 * hour
  if fee < 10.0:
    fee = 10.0
  elif fee > 20.0:
    fee = 20.0

  hour += 1

  print(f"{hour}  {fee}")
```

```python
import math

ticketPrice = 10
defaultAudience = 20
adMoney = 0
print("Ad$\tProfit")
for adMoney in range(0, 201, 25):
  revenue = (defaultAudience + 2 * round(math.sqrt(adMoney))) * ticketPrice
  costs = 200 + adMoney
  profit = revenue - costs
  print(f"{float(adMoney)}\t{float(profit)}")
```

```python
import random

play = input("Do you want to play a game (y/n)?")

while play == "y":
  player = input("Choose between rock, paper, scissors: ").lower()
  computer = random.choice(['rock','paper','scissors'])

  if player == computer:
    print("Draw")
  elif player == "rock" and computer == "scissors":
    print("Player wins!")
  elif player == "scissors" and computer == "paper":
    print("Player wins!")
  elif player == "paper" and computer == "rock":
    print("Player wins!")
  else:
    print("Computer wins!")

play = input("Do you want to play a game (y/n)?")
```

```python
numbers = []

num = int(input('Enter a number: '))

while num != 0:
  numbers.append(num)
  num = int(input('Enter another number: '))
numbers.sort(reverse=True)
print(numbers)
```

```python
numbers = [10.0,9.9,10.1,10.05]

average = sum(numbers) / len(numbers)

largest = numbers[0]
for w in numbers:
  if w > largest:
    largest = w

print("Average =", average)
print("Largest =", largest)
```
